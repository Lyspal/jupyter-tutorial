# Pendulum ODE solver

Solves the pendulum ordinary differential equation numerically:

$$\ddot{\theta}(t) = -\mu \dot{\theta}(t) - \frac{g}{L}\sin{\theta(t)}$$

with initial conditions $\theta_0 = \frac{\pi}{3}$ amd $\dot{\theta}_0 = 0$.

Source: 3Blue1Brown, https://youtu.be/p_di4Zn4wz4

## Creating solver

```python
import numpy as np
import matplotlib.pyplot as plt

# Physical constants
g = 9.8
L = 2
mu = 0.1

# Initial conditions
THETA_0 = np.pi / 3  # 60 degrees
THETA_DOT_0 = 0      # No initial angular velocity

# Definition of ODE
def get_theta_double_dot(theta, theta_dot):
    return -mu * theta_dot - (g / L) * np.sin(theta)

# Solution to the differential equation
def theta(t):
    # Initialize changing values
    theta = THETA_0
    theta_dot = THETA_DOT_0
    delta_t = 0.01  # Some time step
    
    data = [[0.0, theta, theta_dot]]
    
    for time in np.arange(0, t, delta_t):
        theta_double_dot = get_theta_double_dot(theta, theta_dot)
        theta += theta_dot * delta_t
        theta_dot += theta_double_dot * delta_t
        data.append([time, theta, theta_dot])
    
    return data
```

## Retrieving data

```python
d = theta(180)
d = np.array(d)
```

## Plotting data

### Theta and theta_dot

```python
plt.plot(d[:, 1], d[:, 2])
plt.xlabel("theta (radians)")
plt.ylabel("theta dot (rad/s)")
```

### Angle over time

```python
plt.plot(d[:, 0], d[:, 1])
plt.xlabel("time (s)")
plt.ylabel("theta (radians)")
```

### Change in angle over time

```python
plt.plot(d[:, 0], d[:, 2])
plt.xlabel("time (s)")
plt.ylabel("theta dot (rad/s)")
```
